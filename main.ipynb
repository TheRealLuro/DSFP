```python
# STUDENT IDS VERIFIED
invalid_count = 0
seen_entries = []
for index, obj in unclean_csv.iterrows():
    if obj['StudentID'] > 3392 or obj['StudentID'] < 1001:
        print("INVALID ENTRY " + str(obj['StudentID']))
        invalid_count += 1
    elif obj['StudentID'] in seen_entries:
        print("DUPLICATE ENTRY " + str(obj['StudentID']))
        invalid_count += 1
    else:
        seen_entries.append(obj['StudentID'])
print("Complete! Error count: " + str(invalid_count))
```

```python
# AGE VERIFIED
invalid_count = 0
for index, obj in unclean_csv.iterrows():
    if obj['Age'] > 18 or obj['Age'] < 15:
        print("INVALID ENTRY " + str(obj['Age']))
        invalid_count += 1
print("Complete! Error count: " + str(invalid_count))
```

# [GradeClass]
This column is technically "filled", but we shouldn't trust it, since it's rebuildable from the GPA column, which had no NA values. Thus, lets purge it and rebuild.

- 0 = >= 3.5
- 1 = 30.3 - 3.49
- 2 = 2.5-2.99
- 3 = 2.0-2.49
- 4 = <2.0

```python
for index, obj in unclean_csv.iterrows():
    if obj['GPA'] >= 3.5:
        unclean_csv.at[index, 'GradeClass'] = 0
    elif obj['GPA'] >= 3.0:
        unclean_csv.at[index, 'GradeClass'] = 1
    elif obj['GPA'] >= 2.5:
        unclean_csv.at[index, 'GradeClass'] = 2
    elif obj['GPA'] >= 2.0:
        unclean_csv.at[index, 'GradeClass'] = 3
    else:
        unclean_csv.at[index, 'GradeClass'] = 4
unclean_csv['GradeClass'].value_counts()
```

#### Finally, lets condense the database down to proper integer types. While this doesn't matter for saving it, if we began to actually use this dataset, it would be a big deal.

```python
clean_csv = unclean_csv.copy()
```

```python
clean_csv['StudentID'] = clean_csv['StudentID'].astype(np.int16)
clean_csv['Age'] = clean_csv['Age'].astype(np.int8)
clean_csv['Gender'] = clean_csv['Gender'].astype(np.int8)
clean_csv['Absences'] = clean_csv['Absences'].astype(np.int8)
clean_csv['Tutoring'] = clean_csv['Tutoring'].astype(np.int8)
clean_csv['Extracurricular'] = clean_csv['Extracurricular'].astype(np.int8)
clean_csv['Sports'] = clean_csv['Sports'].astype(np.int8)
clean_csv['Music'] = clean_csv['Music'].astype(np.int8)
clean_csv['Volunteering'] = clean_csv['Volunteering'].astype(np.int8)
clean_csv['GradeClass'] = clean_csv['GradeClass'].astype(np.int8)
```

```python
graphing_df = clean_csv.copy()
```

```python
plt.scatter(graphing_df["StudyTimeWeekly"], graphing_df["GPA"], c=graphing_df['Absences'],cmap="viridis", alpha=0.5)
plt.xlabel("Weekly Study Time (hrs)")
plt.ylabel("GPA")
plt.title("Study Time vs GPA vs Absences")
plt.colorbar(label="Absences")
plt.show()
```

```python
graphing_df.groupby("GradeClass")["Absences"].mean().plot(kind="bar")
plt.ylabel("Average Absences")
plt.title("Absences by GPA")
plt.xticks([0, 1, 2, 3, 4], ['>= 3.5', '3.0-3.49', '2.5-2.99', '2.0-2.49', '< 2.0'], rotation=0)
plt.show()
```

```python
for index, obj in unclean_csv.iterrows():
    if obj['Extracurricular'] == 1 or obj['Sports'] == 1 or obj['Music'] == 1 or obj['Volunteering'] == 1:
        unclean_csv.at[index, 'DidAnyExtra'] = 1
    else:
        unclean_csv.at[index, 'DidAnyExtra'] = 0
unclean_csv['DidAnyExtra'].value_counts()
```

```python
unclean_csv.groupby("DidAnyExtra")["GPA"].mean().plot(kind="bar")
plt.xticks([0,1], ["No", "Yes"], rotation=0)
plt.ylabel("Average GPA")
plt.xlabel("Did any Extracurricular Activity")
plt.title("Average GPA by Any Special Activity")
plt.show()
```

```python
order = ["None", "High School", "Some College", "Bachelor's", "Higher"]
unclean_csv.groupby("ParentalEducation")["GPA"].mean().reindex(order).plot(kind="bar")

plt.ylabel("Average GPA")
plt.xlabel("Parents Education Level")
plt.title("Average GPA by Parental Education")
plt.show()
```
